```python
import re

ROW_PATTERN = r'[\.\@]*'
```

## Part 1

```python
with open('inputs/input.txt', 'r') as f:
    grid = []
    count = 0
    for row in f.readlines():
        match = re.search(ROW_PATTERN, row)
        if match:
            grid.append(list(match.group(0)))
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] != '@':
                continue
            check = 0
            for a in range(-1, 2):
                if not (0 <= i+a < len(grid)):
                    continue
                for b in range(-1, 2):
                    row_index = i+a
                    col_index = j+b
                    if a == b == 0 or not (0 <= col_index < len(grid[i])):
                        continue
                    if grid[row_index][col_index]=='@':
                        check += 1
            if check < 4:
                count += 1
    print(count)
                    
```

## Part 2

```python
with open('inputs/input.txt', 'r') as f:
    grid = []
    running_count = 0
    for row in f.readlines():
        match = re.search(ROW_PATTERN, row)
        if match:
            grid.append(list(match.group(0)))
    while True:
        count = 0
        for i in range(len(grid)):
            for j in range(len(grid[i])):
                if grid[i][j] != '@':
                    continue
                check = 0
                for a in range(-1, 2):
                    if not (0 <= i+a < len(grid)):
                        continue
                    for b in range(-1, 2):
                        row_index = i+a
                        col_index = j+b
                        if a == b == 0 or not (0 <= col_index < len(grid[i])):
                            continue
                        if grid[row_index][col_index]=='@':
                            check += 1
                if check < 4:
                    grid[i][j]='x'
                    count+=1
        running_count += count
        if count == 0:
            break
    print(running_count)
```
