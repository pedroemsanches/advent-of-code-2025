## Part 1

```python
lines = []
with open('inputs/input.txt', 'r') as f:
    for line in f:
        line = line.strip()
        lines.append(line)
first_line, second_line, third_line, fourth_line, operations = lines[0].split(), lines[1].split(), lines[2].split(), lines[3].split(), lines[4].split()
width = len(first_line)
total = 0
for col in range(width):
    operation = operations[col]
    string_rep = operation.join([first_line[col], second_line[col], third_line[col], fourth_line[col]])
    total += int(eval(string_rep))
total
```

## Part 2

```python
lines = []
with open('inputs/input.txt', 'r') as f:
    for line in f:
        line = line.strip('\n')
        lines.append(line)
total = 0
intervals = []
for left in [x for x in range(len(lines[-1])) if lines[-1][x]!=' ']:
    right = left+1
    while left <= right and right <= len(lines[-1]):
        if right == len(lines[-1]):
            intervals.append((left, right))
            break
        elif lines[-1][right] in ['+', '*']:
            intervals.append((left, right-1))
            break
        right += 1

for interval in intervals:
    if lines[-1][interval[0]]=='*':
        sum = 1
    else:
        sum = 0
    for i in reversed(range(*interval)):
        n = int(lines[0][i]+lines[1][i]+lines[2][i]+lines[3][i])
        if lines[-1][interval[0]]=='*':
            sum *= n
        else:
            sum += n
    total += sum

total
```
