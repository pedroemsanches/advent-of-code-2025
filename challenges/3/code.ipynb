## Part 1

```python
with open('inputs/input.txt', 'r') as f:
    banks = f.readlines()
    count = 0
    for bank in banks:
        bank = bank.split('\n')[0]
        len_bank = len(bank)
        max_number_a = max_number_b = 0
        max_index = 0
        for i in range(len_bank-1):
            number_a = int(bank[i])
            if number_a > max_number_a:
                max_number_a = number_a
                max_index = i
        for j in range(max_index+1, len_bank):
            number_b = int(bank[j])
            if number_b > max_number_b:
                max_number_b = number_b

        count += (max_number_a*10)+max_number_b
    print(count)
```

## Part 2

```python
def largest_k_seq(k: int, bank: str):
    to_remove = len(bank)-k
    stack = []

    for digit in bank:
        while to_remove > 0 and len(stack)>0 and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    return ''.join(stack[:k])

with open('inputs/input.txt', 'r') as f:
    banks = f.readlines()
    count = 0
    for bank in banks:
        bank = bank.split('\n')[0]
        largest = largest_k_seq(12, bank)
        count+=int(largest)
    print(count)
```
