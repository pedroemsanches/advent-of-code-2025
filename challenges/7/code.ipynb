## Part 1

```python
lines = []
with open('inputs/input.txt', 'r') as f:
    for line in f:
        line = line.strip()
        lines.append(line)
curr = set([lines[0].index('S')])
count = 0
for i in range(1, len(lines)):
    to_be_splitted = []
    for j in curr:
        if lines[i][j]=='^':
            count+=1
            to_be_splitted.append(j)
    for item in to_be_splitted:
        curr.remove(item)
        curr.update([item-1, item+1])
print(count)
```

## Part 2

```python
lines = []
with open('inputs/input.txt', 'r') as f:
    for line in f:
        line = line.strip()
        lines.append(line)
paths = [0]*len(lines[0])
paths[lines[0].index('S')]=1
for i in range(1, len(lines)):
    to_be_splitted = []
    for j in range(len(lines[i])):
        if lines[i][j]=='^':
            to_be_splitted.append(j)
    for item in to_be_splitted:
        paths[item+1] += paths[item]
        paths[item-1] += paths[item]
        paths[item] = 0
sum(paths)
```
