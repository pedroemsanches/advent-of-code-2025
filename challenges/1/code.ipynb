## Part 1

```python
with open('inputs/input.txt', 'r') as f:
    curr = 50
    count = 0
    for line in f.readlines():
        instruction = line.split('\n')[0]
        mult = 1
        if instruction[0]=='L':
            mult = -1
        curr = curr + mult*int(instruction[1:])
        if curr%100 == 0:
            count += 1
    print(count)
```

## Part 2

```python
with open('inputs/input.txt', 'r') as f:
    curr = next = 50
    count = 0
    for line in f.readlines():
        instruction = line.split('\n')[0]
        mult = 1
        if instruction[0]=='L':
            mult = -1
        next = curr + mult*int(instruction[1:])
        curr_turn = curr // 100
        next_turn = next // 100

        diff = abs(curr_turn-next_turn)

        if curr % 100 == 0 and mult == -1:
            diff-=1
        if next % 100 == 0 and mult == -1:
            diff+=1
        count += diff
        curr = next % 100
    print(count)
```
