## Part 1

```python
with open('inputs/input.txt', 'r') as f:
    ranges = f.readline().split('\n')[0].split(',')
    count = 0
    for r in ranges:
        initial, final = r.split('-')[0], r.split('-')[1]
        len_init, len_final = len(initial), len(final)
        if len_init == len_final and len_init%2 == 1:
            continue
        for n in range(int(initial), int(final)+1):
            str_n = str(n)
            length = len(str_n)
            first_part, last_part = str_n[:length//2], str_n[length//2:]
            if first_part == last_part:
                count += n
    print(count)
```

## Part 2

```python
with open('inputs/input.txt', 'r') as f:
    ranges = f.readline().split('\n')[0].split(',')
    count = 0
    for r in ranges:
        initial, final = r.split('-')[0], r.split('-')[1]
        for n in range(int(initial), int(final)+1):
            str_n = str(n)
            length = len(str_n)
            flag = False
            for i in range(1, length):
                if length % i != 0:
                    continue
                comp = set([str_n[j:j+i] for j in range(0, length, i)])
                # print(r, n, comp)
                if len(comp) == 1:
                    flag = True
            if flag:
                count+=n
    print(count)
```
